```python

# Import the userdata module from google.colab
from google.colab import userdata
import os

# Retrieve secrets and set environment variables
os.environ["KAGGLE_USERNAME"] = "your_username"
os.environ["Kaggle_Key"] = "your_api_key"

print("Kaggle API keys loaded from Colab Secrets.")

# You can then list datasets as usual
!pip install -q kaggle
!kaggle datasets list

#Laster ned spesifikt datasett
!kaggle datasets download -d masoudnickparvar/brain-tumor-mri-dataset -p /content --unzip

#Verifiserer mappelagring.
!ls -la /content
!find /content -maxdepth 2 -type d -name "Training" -o -name "Testing"

#Initierer datastruktur
from pathlib import Path
import random, shutil

random.seed(42)

src_train = Path("/content/Training")
src_test  = Path("/content/Testing")

dst_root  = Path("/content/data")
dst_train = dst_root / "train"
dst_val   = dst_root / "val"
dst_test  = dst_root / "test"

if dst_root.exists():
    shutil.rmtree(dst_root)

shutil.copytree(src_test, dst_test)

# Create train/val split
VAL_FRAC = 0.15
classes = [d.name for d in src_train.iterdir() if d.is_dir()]
print("Classes found:", classes)

for c in classes:
    imgs = sorted((src_train / c).glob("*"))
    imgs = [p for p in imgs if p.is_file()]
    random.shuffle(imgs)

    n_val = int(len(imgs) * VAL_FRAC)

    (dst_train / c).mkdir(parents=True, exist_ok=True)
    (dst_val / c).mkdir(parents=True, exist_ok=True)

    for p in imgs[:n_val]:
        shutil.copy2(p, dst_val / c / p.name)

    for p in imgs[n_val:]:
        shutil.copy2(p, dst_train / c / p.name)

print("Split complete.")

#Sjekker datasettfordeling
from pathlib import Path

for split in ["train", "val", "test"]:
    print("\n", split.upper())
    base = Path("/content/data") / split
    for c in sorted([d.name for d in base.iterdir() if d.is_dir()]):
        print(c, ":", len(list((base/c).glob("*"))))

#Laster data for trening
import tensorflow as tf
import os
print("TF version:", tf.__version__)
print("GPU:", tf.config.list_physical_devices('GPU'))

IMG_SIZE = (256, 256)
BATCH = 32
SEED = 42

train_dir = "/content/data/train"
val_dir = "/content/data/val"
test_dir = "/content/data/test"

train_ds = tf.keras.utils.image_dataset_from_directory(
    train_dir,
    label_mode = "categorical",
    image_size = IMG_SIZE,
    batch_size = BATCH,
    shuffle = True,
    seed = SEED
)

val_ds = tf.keras.utils.image_dataset_from_directory(
    val_dir,
    label_mode="categorical",
    image_size=IMG_SIZE,
    batch_size=BATCH,
    shuffle=False
)
test_ds = tf.keras.utils.image_dataset_from_directory(
    test_dir,
    label_mode="categorical",
    image_size=IMG_SIZE,
    batch_size=BATCH,
    shuffle=False
)
class_names = train_ds.class_names
num_classes = len(class_names)
print("Classes:", class_names, "num_classes", num_classes)

AUTOTUNE = tf.data.AUTOTUNE
train_ds = train_ds.prefetch(AUTOTUNE)
val_ds = val_ds.prefetch(AUTOTUNE)
test_ds = test_ds.prefetch(AUTOTUNE)

#Definerer data-augmentering
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal"),
    tf.keras.layers.RandomRotation(0.05),
    tf.keras.layers.RandomZoom(0.1),
    tf.keras.layers.RandomContrast(0.2),
])

#Bygger og kompilerer ResNet50-modell
from tensorflow.keras import layers, models

base = tf.keras.applications.ResNet50(
    include_top=False,
    weights="imagenet",
    input_shape=IMG_SIZE + (3,)
)
base.trainable = True  #To freeze or not freeze
for layer in base.layers[:-50]: #how much layers are to unfreeze
    layer.trainable = False

inputs = layers.Input(shape=IMG_SIZE + (3,))
x = data_augmentation(inputs)
x = tf.keras.applications.resnet.preprocess_input(x)
x = base(x, training=True)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dropout(0.3)(x)
outputs = layers.Dense(num_classes, activation="softmax")(x)

model = models.Model(inputs, outputs)

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

model.summary()

#Beregner klassevekter
import numpy as np
!pip -q install scikit-learn

import numpy as np
from sklearn.utils.class_weight import compute_class_weight

y_train = np.concatenate([np.argmax(y.numpy(), axis=1) for x,y in train_ds], axis=0)
weights = compute_class_weight("balanced", classes=np.unique(y_train), y=y_train)
class_weight = {i: w for i, w in enumerate(weights)}
print(class_weight)

#Trener modell med tilbakekallinger
callbacks = [
    tf.keras.callbacks.ModelCheckpoint(
        "resnet50_best.keras",
        monitor="val_accuracy",
        save_best_only=True,
        mode="max"
    ),
    tf.keras.callbacks.EarlyStopping(
        monitor="val_accuracy",
        patience=3,
        restore_best_weights=True
    ),
    tf.keras.callbacks.ReduceLROnPlateau(
        monitor="val_loss",
        factor=0.5,
        patience=2
    ),
]

history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=30,
    callbacks=callbacks,
    class_weight=class_weight
)


best = tf.keras.models.load_model("resnet50_best.keras")
test_loss, test_acc = best.evaluate(test_ds)
print("Best checkpoint test acc:", test_acc)

y_pred_probs = best.predict(test_ds)

#Evaluerer modell på testdata
test_loss, test_acc = best.evaluate(test_ds)
print("Test acc:", test_acc)

#Viser forvirringsmatrise/rapport
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report

# Get y_true and y_pred
y_true = np.concatenate([np.argmax(y.numpy(), axis=1) for x, y in test_ds], axis=0)
y_pred_probs = best.predict(test_ds)
y_pred = np.argmax(y_pred_probs, axis=1)

cm = confusion_matrix(y_true, y_pred)
print("Confusion Matrix:\n", cm)

print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=class_names, digits=4))

#Visualiserer forvirringsmatrise
import seaborn as sns
import matplotlib.pyplot as plt
import json


plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt="d",
            xticklabels=class_names,
            yticklabels=class_names,
            cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history["loss"], label="Train loss")
plt.plot(history.history["val_loss"], label="Val loss")
plt.legend(); plt.title("Loss")

plt.subplot(1, 2, 2)
plt.plot(history.history["accuracy"], label="Train acc")
plt.plot(history.history["val_accuracy"], label="Val acc")
plt.legend(); plt.title("Accuracy")
plt.show()

config = {
    "image_size": [256, 256],
    "classes": class_names,
    "preprocessing": "resnet"
}

with open("config.json", "w") as f:
    json.dump(config, f, indent=2)

best.save("resnet50_best.keras")

with open("class_names.json", "w") as f:
    json.dump(class_names, f)

import os
print(os.listdir())

from google.colab import files

files.download("resnet50_best.keras")
files.download("class_names.json")
files.download("config.json")
```
